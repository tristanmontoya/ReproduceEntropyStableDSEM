# Nodal sets on the triangle

```julia
using StableSpectralElements, Plots, LaTeXStrings
```

```julia
ref_tri_multi_p4 = ReferenceApproximation(
    NodalMulti(4), Tri(), mapping_degree=3)
p_tri_multi_p4 = plot(windowsize=(350,300),ref_tri_multi_p4, sketch=false, volume_quadrature=true,
    volume_node_color=:slateblue4,
    facet_quadrature=false,
    node_color=:slateblue4,
    fontfamily="Computer Modern",
    markersize=6,xtickfontsize=12, ytickfontsize=12, 
    xguidefontsize=14, yguidefontsize=14,
    xlabel=string("\n",L"\xi{}_1"), ylabel=string(L"$\xi{}_2$","\n"),
    xlims=[-1.1,1.1],ylims=[-1.1,1.1],
    xticks=([-1.0,0.0,1.0], [L"-$\,1.0$",L"0.0",L"1.0"]), 
    yticks=([-1.0,0.0,1.0],[L"-$\,1.0$",L"0.0",L"1.0"]),
    grid_line_width=3, edge_line_width=5)
```

```julia
ref_tri_tensor_p4 = ReferenceApproximation(
    NodalTensor(4), Tri(), mapping_degree=3)
p_tri_tensor_p4 = plot(windowsize=(350,300),
    ref_tri_tensor_p4, sketch=false, volume_quadrature=true,
    grid_connect=true,
    node_color=palette(:default)[2],
    facet_quadrature=false,
    fontfamily="Computer Modern",
    markersize=6,xtickfontsize=12, ytickfontsize=12, 
    xguidefontsize=14, yguidefontsize=14,
    xlabel=string("\n",L"\xi{}_1"), ylabel=string(L"$\xi{}_2$","\n"),
    xlims=[-1.1,1.1],ylims=[-1.1,1.1],
    xticks=([-1.0,0.0,1.0], [L"-$\,1.0$",L"0.0",L"1.0"]), 
    yticks=([-1.0,0.0,1.0],[L"-$\,1.0$",L"0.0",L"1.0"]),
    grid_line_width=3, edge_line_width=5)
scatter!(p_tri_tensor_p4, ref_tri_tensor_p4.reference_element.rq, ref_tri_tensor_p4.reference_element.sq,
    markershape=:circle, markersize=6, markercolor=palette(:default)[2], markerstrokewidth=0.0)
```

```julia
savefig(p_tri_multi_p4, "../plots/multi_nodes_p4_tri.pdf")
savefig(p_tri_tensor_p4, "../plots/tensor_nodes_p4_tri.pdf")
```
