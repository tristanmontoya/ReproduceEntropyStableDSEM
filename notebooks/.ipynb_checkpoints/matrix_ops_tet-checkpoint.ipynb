```julia
using StableSpectralElements, GFlops, LinearAlgebra, Plots, LaTeXStrings
```

```julia
p = [i for i in 2:15]

function N_ops_tensor(ref_approx)
    V = count_ops(make_operator(ref_approx.V, DefaultOperatorAlgorithm()))
    V_trans = count_ops(make_operator(ref_approx.V', DefaultOperatorAlgorithm()))
    R = count_ops(make_operator(ref_approx.R, GenericTensorProductAlgorithm()))
    R_trans = count_ops(make_operator(ref_approx.R, GenericTensorProductAlgorithm())')
    
    conservative_variables = V + R
    entropy_projection = 2*ref_approx.N_q + 2*V_trans + V
    entropy_variables = V + R
    facet_operators = R_trans + ref_approx.N_f + ref_approx.N_q 
    # multiply by facet quad weights and add to volume nodes
    time_derivative = ref_approx.N_q + 2*V_trans + V
    
    return conservative_variables + entropy_projection + entropy_variables + facet_operators + time_derivative
end

ops_tensor = zeros(14)

print("p = ")
for i in eachindex(p)
    print(p[i], ",")
    approx_type = ModalTensor(p[i])
    ref_approx = ReferenceApproximation(approx_type, Tet())
    ops_tensor[i] = N_ops_tensor(ref_approx)
end
```

```julia
p_multi = [i for i in 2:10]

function N_ops_multi(ref_approx)
    V = count_ops(make_operator(ref_approx.V, GenericMatrixAlgorithm()))
    V_trans = count_ops(make_operator(ref_approx.V', GenericMatrixAlgorithm()))
    R = count_ops(make_operator(ref_approx.R, GenericMatrixAlgorithm()))
    R_trans = count_ops(make_operator(ref_approx.R, GenericMatrixAlgorithm())')
    
    conservative_variables = V + R
    entropy_projection = 2*ref_approx.N_q + 2*V_trans + V
    entropy_variables = V + R
    facet_operators = R_trans + ref_approx.N_f + ref_approx.N_q # multiply by facet quad weights and add to volume nodes
    time_derivative = ref_approx.N_q + 2*V_trans + V
    
    return conservative_variables + entropy_projection + entropy_variables + facet_operators + time_derivative
end

ops_multi = zeros(9)

print("p = ")
for i in eachindex(p[1:9])
    print(p[i], ",")
    approx_type = ModalMulti(p[i])
    ref_approx = ReferenceApproximation(approx_type, Tet())
    ops_multi[i] = N_ops_multi(ref_approx)
end
```

```julia
plt = plot(legend=:topleft, xlabel="\n Polynomial degree \$p\$", 
    xscale=:log10, 
    yscale=:log10, 
    ylims=[1000,10^7], 
    #xlims=[1,16], ylims=[1,6],
    ylabel="Floating-point operations\n", fontfamily="Computer Modern", legendfontsize=10,
    xguidefontsize=12, yguidefontsize=12, xtickfontsize=12, ytickfontsize=12,
    xticks=([2,5,10,15], string.([2,5,10,15])), size=(450,300))
plot!(plt, p_multi, ops_multi, label="Multidimensional", 
    linestyle=:solid, markershape=:circle, markersize=5, linewidth=2, 
    markerstrokewidth=0.0, color=:slateblue4)
plot!(plt, p, ops_tensor, label="Tensor-product", 
    linestyle=:solid, markershape=:square, markersize=3, linewidth=2, 
    markerstrokewidth=0.0, color=2)

r1 = 12
r2 = 15
plot!(plt, [r1, r2, r2, r1], 15*[r1^4, r2^4, r1^4, r1^4], label=:none, 
    linewidth=2, color=:black, annotations = ([13.7], [2.2*10^5], L"\mathbf{4:1}", 10))
r1 = 8
r2 = 10
plot!(plt, [r1, r1, r2, r1], 3.1*[r1^6, r2^6, r2^6, r1^6], label=:none, 
    linewidth=2, color=:black, annotations = ([9], [4.4*10^6], L"\mathbf{6:1}", 10))
```
