# Collapsed coordinate transformation in 2D

```julia
using StableSpectralElements, Plots, LaTeXStrings, Plots.PlotMeasures
p = 6
ref_tri = ReferenceApproximation(
    NodalTensor(p), Tri(), volume_quadrature_rule=Tuple(LGLQuadrature(p) for m in 1:2));
ref_quad = ReferenceApproximation(
    NodalTensor(p), Quad(), volume_quadrature_rule=LGLQuadrature(p));
```

```julia
tri = plot(windowsize=(350,300),
    ref_tri, sketch=false, volume_quadrature=true,
    grid_connect=true,
    node_color=:grey25,
    facet_quadrature=false,
    fontfamily="Computer Modern",
    markersize=6,xtickfontsize=12, ytickfontsize=12, 
    xguidefontsize=14, yguidefontsize=14,
    xlabel=string("\n",L"\xi{}_1"), ylabel=string(L"$\xi{}_2$","\n"),
    xlims=[-1.1,1.1],ylims=[-1.1,1.1],
    xticks=([-1.0,0.0,1.0], [L"-$\,1.0$",L"0.0",L"1.0"]), 
    yticks=([-1.0,0.0,1.0],[L"-$\,1.0$",L"0.0",L"1.0"]),
    grid_line_width=3, edge_line_width=0)
```

```julia
savefig(tri, "../plots/tri_coords.pdf")
```

```julia
quad = plot(windowsize=(350,300),
    ref_quad, sketch=false, volume_quadrature=true,
    grid_connect=true,
    node_color=:grey25,
    facet_quadrature=false,
    fontfamily="Computer Modern",
    markersize=6,xtickfontsize=12, ytickfontsize=12, 
    xguidefontsize=14, yguidefontsize=14,
    xlabel=string("\n",L"\eta{}_1"), ylabel=string(L"$\eta{}_2$","\n"),
    xlims=[-1.1,1.1],ylims=[-1.1,1.1],
    xticks=([-1.0,0.0,1.0], [L"-$\,1.0$",L"0.0",L"1.0"]), 
    yticks=([-1.0,0.0,1.0],[L"-$\,1.0$",L"0.0",L"1.0"]),
    grid_line_width=3, edge_line_width=0)
```

```julia
savefig(quad, "../plots/quad_coords.pdf")
```
