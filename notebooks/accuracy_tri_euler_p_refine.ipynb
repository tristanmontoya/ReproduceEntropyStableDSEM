# Two-dimensional density wave

```julia
using JLD2, Plots, LaTeXStrings
```

```julia
path = "../results/euler_p_refine/"

p_mm_ec = load_object(string(path, 
        "ModalMulti_Tri_FluxDifferencingForm/EntropyConservativeNumericalFlux/poly_degrees.jld2"))
p_mm_lf = load_object(string(path, 
        "ModalMulti_Tri_FluxDifferencingForm/LaxFriedrichsNumericalFlux/poly_degrees.jld2"))
p_mt_ec = load_object(string(path, 
        "ModalTensor_Tri_FluxDifferencingForm/EntropyConservativeNumericalFlux/poly_degrees.jld2"))
p_mt_lf = load_object(string(path, 
        "ModalTensor_Tri_FluxDifferencingForm/LaxFriedrichsNumericalFlux/poly_degrees.jld2"))

e_mm_ec = load_object(string(path, 
        "ModalMulti_Tri_FluxDifferencingForm/EntropyConservativeNumericalFlux/errors.jld2"))
e_mm_lf = load_object(string(path, 
        "ModalMulti_Tri_FluxDifferencingForm/LaxFriedrichsNumericalFlux/errors.jld2"))
e_mt_ec = load_object(string(path, 
        "ModalTensor_Tri_FluxDifferencingForm/EntropyConservativeNumericalFlux/errors.jld2"))
e_mt_lf = load_object(string(path, 
        "ModalTensor_Tri_FluxDifferencingForm/LaxFriedrichsNumericalFlux/errors.jld2"));
p_min = 2
p_max=15;
```

```julia
e = 1
p_conv = plot(yaxis=:log10, windowsize=(350,300), legendfontsize=9,
    fontfamily="Computer Modern", xtickfontsize=12, ytickfontsize=12, 
    xguidefontsize=12, yguidefontsize=12, ylabel="\$L^2\$ error\n", 
    xlabel="\nPolynomial degree \$p\$", legend=:topright,  ylims=[1e-6,10e-1], 
    yticks=([1e-6, 1e-4, 1e-2, 1e0], [L"10^{-6}", L"10^{-4}", L"10^{-2}", L"10^{0}"]),
    xticks=([5,10,15], [L"5", L"10", L"15"]))
first=p_min-1
last=p_max-1

plot!(p_conv, collect(p_min:p_max),[e_mm_ec[i][e] for i in first:last],  label="", 
    linewidth=2.0, linestyle=:dot, markershape=:circle, 
    markersize=6, markerstrokewidth=0.0, markerstrokecolor=:slateblue4, 
    color=:slateblue4)
plot!(p_conv, collect(p_min:p_max), [e_mm_lf[i][e] for i in first:last], 
    label="Multidimensional", 
    linewidth=2.0, linestyle=:solid, markershape=:circle, markersize=6, 
    markerstrokewidth=0.0, markerstrokecolor=:slateblue4, color=:slateblue4)
plot!(p_conv, collect(p_min:p_max),  [e_mt_ec[i][e] for i in first:last], 
    label="", 
    linewidth=2.0, linestyle=:dot, markershape=:square, markersize=3, 
    markerstrokewidth=0.0, color=palette(:default)[2])
plot!(p_conv, collect(p_min:p_max), [e_mt_lf[i][e] for i in first:last], label="Tensor-product", 
    linewidth=2.0, linestyle=:solid, markershape=:square, markersize=3,
    markerstrokewidth=0.0, color=palette(:default)[2])
```

```julia
savefig(p_conv, "../plots/euler_p_refine_tri.pdf")
```
